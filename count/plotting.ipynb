```python
import json
import matplotlib.pyplot as plt
import seaborn
```

```python
with open('barcode_counts_n70.json') as json_file:
    n70 = json.load(json_file)

with open('barcode_counts_n71.json') as json_file:
    n71 = json.load(json_file)

with open('barcode_counts_n72.json') as json_file:
    n72 = json.load(json_file)
```

```python
n70_counts = sorted(list(n70.values()), reverse=True)
n71_counts = sorted(list(n71.values()), reverse=True)
n72_counts = sorted(list(n72.values()), reverse=True)
```

```python
x_n70 = n70_counts[0:15000]
y_n70 = range(len(x_n70))

plt.bar(y_n70, x_n70, color='black', width=75)
plt.title('N70 Barcode')
plt.ylabel('Reads')
plt.xticks([])

plt.savefig('N70_barcode.png')
```

```python
x_n71 = n71_counts[0:200]
y_n71 = range(len(x_n71))

plt.bar(y_n71, x_n71, color='black', width=1)
plt.title('N71 Barcode')
plt.ylabel('Reads')
plt.xticks([])

plt.savefig('N71_barcode.png')
```

```python
x_n72 = n72_counts[0:1250]
y_n72 = range(len(x_n72))

plt.bar(y_n72, x_n72, color='black', width=10)
plt.title('N72 Barcode')
plt.ylabel('Reads')
plt.xticks([])

plt.savefig('N72_barcode.png')
```
