```python
import json
import matplotlib.pyplot as plt
import seaborn
```

```python
def json_reader(file, n):
    """
    Reads a json file and returns a list of the top n most frequent words in the file
    
    :param file: the file name of the json file
    :param n: number of top articles to return
    :return: A list of the top n most frequent words in the file.
    """
    with open(file, 'r') as json_file:
        x = json.load(json_file)

    sortx = {k: v for k, v in sorted(x.items(), key=lambda item: item[1], reverse= True)}
    return list(sortx.values())[:n]
```

```python
forward_n70 = json_reader('unprocessed_data/forward_barcode_counts_n70.json', 15000)
y_n70 = range(len(forward_n70))

plt.bar(y_n70, forward_n70, color='black', width=75)
plt.title('N70 Barcode Counts in Forward Read')
plt.ylabel('Reads')
plt.xticks([])

plt.savefig('/home/eck/barcode/results/N70_forward_barcode.png')
```
